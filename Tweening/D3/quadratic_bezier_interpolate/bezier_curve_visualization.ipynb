```python
import matplotlib.pyplot as plt

def cubic_bezier(t, p1, p2):
    cp1x, cp1y = p1
    cp2x, cp2y = p2
    x = 3 * (1 - t)**2 * t * cp1x + 3 * (1 - t) * t**2 * cp2x + t**3
    y = 3 * (1 - t)**2 * t * cp1y + 3 * (1 - t) * t**2 * cp2y + t**3
    return x, y

# Generate 10 frames for the cubic Bezier curve
frames = []
for i in range(11):  # More points for a smoother curve
    t = i / 10
    point = cubic_bezier(t, (0.42, 0), (0.58, 1))  # easeInOutCubic control points
    frames.append((point[0], point[1]))

# Unpack points for plotting
x_vals, y_vals = zip(*frames)

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, label='Cubic Bezier Curve', marker='o')
plt.title('Cubic Bezier Curve Representation')
plt.xlabel('Normalized Time (t)')
plt.ylabel('Eased Value')
plt.grid(True)
plt.legend()
plt.show()
```
